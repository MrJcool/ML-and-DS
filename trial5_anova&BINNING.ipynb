```python
import pandas as pd
import numpy as np
raw41 = pd.read_csv('transformed_housing_data')
```

```python
#in python variable neme cannot have spaces in between so we replace the variable names of categorical variables 
raw41 = raw41.rename(columns = {'Sale Price' : 'Sale_Price'})
```

```python
raw41.dtypes
```

```python
raw41 = raw41.rename(columns = {'Waterfront View' : 'Waterfront_View'})
raw41 = raw41.rename(columns = {'Condition of the House' : 'Condition_of_the_House'})
raw41 = raw41.rename(columns = {'Ever Renovated' : 'Ever_Renovated'})
```

```python
raw41['Zipcode'] = raw41['Zipcode'].astype(object)
```

# 

```python
raw41.dtypes
```

```python
raw41.head()
```

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm
```

```python
mod = ols('Sale_Price ~ Condition_of_the_House',data = raw41).fit()
```

```python
anova_table = sm.stats.anova_lm(mod, typ =2)
```

```python
anova_table
```

```python
mod = ols('Sale_Price ~ Zipcode',data = raw41).fit()
sm.stats.anova_lm(mod, typ =2)
```

```python
mod = ols('Sale_Price ~ Waterfront_View',data = raw41).fit()
sm.stats.anova_lm(mod, typ =2)
```

```python
mod = ols('Sale_Price ~ Ever_Renovated',data = raw41).fit()
sm.stats.anova_lm(mod, typ =2)
```

# dummy variables and BINNING

```python
raw41 = pd.get_dummies(raw41, columns = ['Condition_of_the_House'], drop_first = True)
```

```python
raw41 = pd.get_dummies(raw41, columns = ['Waterfront_View'], drop_first = True)
raw41 = pd.get_dummies(raw41, columns = ['Ever_Renovated'], drop_first = True)
```

# 

```python
raw41
```

```python
zip_table = raw41.groupby('Zipcode').agg({'Sale_Price' : 'mean'}).sort_values('Sale_Price', ascending = True)
```

```python
zip_table
```

```python
zip_table['Zipcode_Group'] = pd.cut(zip_table['Sale_Price'], bins = 10, 
                  labels = ['Zipcode_Group_0',
                            'Zipcode_Group_1',
                             'Zipcode_Group_2','Zipcode_Group_3',
                  'Zipcode_Group_4','Zipcode_Group_5','Zipcode_Group_6',
                  'Zipcode_Group_7','Zipcode_Group_8','Zipcode_Group_9'], include_lowest = True)
```

```python
zip_table = zip_table.drop(columns = 'Sale_Price')
zip_table
```

```python
#using merge function and left_join
raw41 = pd.merge(raw41, zip_table,  left_on = 'Zipcode', how = 'left', right_index = True)
```

```python
raw41 = raw41.drop(columns = 'Zipcode')
```

```python
raw41.head()
```

```python
raw41 = pd.get_dummies(raw41, columns = ['Zipcode_Group'], drop_first = True)
```
