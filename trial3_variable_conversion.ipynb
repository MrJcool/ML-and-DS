```python
import pandas as pd
import numpy as np
raw31 = pd.read_csv('raw3')
raw31['Zipcode'] = raw31['Zipcode'].astype(object)
raw31
```

```python
raw31['No of Times Visited'].unique()
mapping = {'None' : '0','Once' : '1','Twice' : '2','Thrice' : '3','Four' : '4'}
raw31['No of Times Visited'] = raw31['No of Times Visited'].map(mapping)
#raw31['No of Times Visited'].unique()
```

```python
raw31['No of Times Visited'] = raw31['No of Times Visited'].astype(int)
raw31['No of Times Visited'].unique()
```

```python
#from the variablr "Renovated Year" two variables are formed  to derive useful info from the data 
raw31['Ever Renovated'] = np.where(raw31['Renovated Year'] == 0,"No", 'Yes')
raw31
```

```python
raw31['Purchase Year'] = pd.DatetimeIndex(raw31['Date House was Sold']).year
raw31
```

```python
raw31['Year since Renovation'] = np.where(raw31['Ever Renovated'] == 'Yes',abs(raw31['Purchase Year'] - raw31['Renovated Year']),0)
raw31
```

```python
raw31.drop(columns = ['Date House was Sold','Renovated Year','Purchase Year'], inplace = True)
raw31
```

```python
raw31.to_csv('transformed_housing_data',index = False)
```
