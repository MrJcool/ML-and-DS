```python
def adj_R2(data_x, data_y, predictions):
    from sklearn.metrics import r2_score
    R = r2_score(data_y, predictions)
    k = len(data_x.columns)
    n = len(data_y)
    adj_R = 1 - ((n -1)/(n - k - 1)) * (1 - R)
    return adj_R
```

# Feature Selection

```python
def adj_r2(m, data_y, predictions):
    from sklearn.metrics import r2_score
    R = r2_score(data_y, predictions)
    n = len(data_y)
    adj_R = 1 - ((n -1)/(n - m - 1)) * (1 - R)
    return adj_R
```

```python
def return_r2(X, Y, model):
    '''
    X = independent variable
    Y = Independent variable / target
    model = predictive model
    
    returns the adjusted R2
    '''
    model.fit(X, Y)
    pred = model.predict(X)
    m = len(X.columns)
    return adj_r2(m, Y, pred)
```

```python
def forward_selection(model, X, Y, max_features):
    '''
    X = independent variable
    Y = Independent variable / target
    model = predictive model
    max_features = selects best features upto max features value
    
    returns
    Rmax = adjusted r2 of the final model
    history = list of Rmax for every independent variable added to the variable
    X[f_col] = best features upto max features
    '''
    f_col = []
    history = []
    flag = 0
    elim = 0
    for i in range(max_features):
        f_col.append('blank')
        #print(f_col)
        tmp = f_col[:]
        Rmax = -1
        for var in X.columns:
            if var not in tmp:
                tmp[-1] = var
                r2 = return_r2(X[tmp], Y, model)
                flag += 1
                #print(tmp)
                if r2 >= Rmax:
                    f_col[-1] = var
                    Rmax = r2
                    elim = var
            #X.drop(columns = [elim], inplace = True)
        print('Selected', f_col[-1], 'with new R2', Rmax,'---',flag)
        history.append(Rmax)
    return Rmax, history, X[f_col]
```
