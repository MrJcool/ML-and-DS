```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.simplefilter('ignore')
plt.style.use('dark_background')
from sklearn.metrics import mean_squared_error as mse
```

```python
data = pd.read_csv('raw3')
data.head()
```

```python
data.info()
```

```python
num_data = data.drop(columns = ['ID', 'Date House was Sold', 'Sale Price', 'Waterfront View', 'No of Times Visited', 
                               'Condition of the House', 'Overall Grade', 'Zipcode'])
num_data.info()
```

```python
from sklearn.preprocessing import Normalizer
norm = Normalizer()
norm_data = norm.fit_transform(num_data)

norm_data = pd.DataFrame(norm_data, columns = num_data.columns)
norm_data
```

```python
tmp = norm_data.var()
tmp
```

```python
T = 1e-04

for i in tmp:
    if i < T:
        norm_data.drop(columns = tmp[tmp == i].index[0], inplace = True)
norm_data.var()
```
