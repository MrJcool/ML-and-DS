# Missing value Ratio Technique

```python
data = pd.read_csv('NaN_Housing_Prices.csv')
data
```

```python
data.isnull().sum()
```

```python
for i in data.columns:
    tmp = data[i].isnull().sum() / len(data)
    if tmp > 0.7:
        print("Removing the '" + i + "' column from the dataset" , tmp)
        data = data.drop(columns = [i])
```
