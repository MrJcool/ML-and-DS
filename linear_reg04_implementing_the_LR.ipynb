# Scaling the dataset

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
Y = data['Sale_Price']
X = scaler.fit_transform(data.drop(columns = ['Sale_Price']))
X = pd.DataFrame(data = X, columns = data.drop(columns = ['Sale_Price']).columns)
X.head()
```

# Checking for Multicollinearity and removing it

```python
X.corr()
```

```python
## finding the independent variables with collinearityh greater than 0.5 
k = X.corr()
z = [[str(i), str(j)] for i in k.columns for j in k.columns if(k.loc[i,j] > abs(0.5)) & (i != j)]
z, len(z)
```

# Calculating the VIF

```python
## importing the variance_inflation_factor from the statsmodel library
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif

vif_data = X

VIF = pd.Series([vif(vif_data.values, i) for i in range(vif_data.shape[1])], index = vif_data.columns)
VIF
```

```python
VIF[VIF == VIF.max()].index[0]
```

```python
def MC_remover(data):
    vif1 = pd.Series([vif(data.values, i) for i in range(data.shape[1])], index = data.columns)
    if vif1.max() > 5:
        print(vif1[vif1 == vif1.max()].index[0], 'has been removed')
        data = data.drop(columns = [vif1[vif1 == vif1.max()].index[0]])
        return data
    else:
        print('No multicollinearity present anymore')
        return data
```

```python
for i in range(7):
    vif_data = MC_remover(vif_data)

vif_data.head()
```

# Calculating the VIF of remaining columns

```python
VIF = pd.Series([vif(vif_data.values, i) for i in range(vif_data.shape[1])], index = vif_data.columns)
VIF
```

# Train/Test Dataset

```python
X = vif_data
Y = data['Sale_Price']
```

```python
from sklearn.model_selection import train_test_split as tts
x_train, x_test, y_train, y_test = tts(X, Y, test_size = 0.3, random_state = 101)

x_test.shape, x_train.shape, y_test.shape, y_train.shape
```

# implementing linear regression using sklearn library

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression(normalize = True)
lr.fit(x_train, y_train)
```

```python
lr.coef_
```

```python
predictions = lr.predict(x_test)
lr.score(x_test, y_test)
```

# 1. Residuals

```python
residual = predictions - y_test

residual_table = pd.DataFrame({'Residuals': residual, 'Predictions': predictions}
                             )
 
residual_table = residual_table.sort_values(by = 'Predictions') 
```

```python
z = [i for i in range(int(residual_table['Predictions'].max()))]
k = [0 for i in range(int(residual_table['Predictions'].max()))]
```

```python
plt.figure(dpi = 150, figsize = (17, 7))
plt.scatter(residual_table['Predictions'], residual_table['Residuals'], color = 'red', s = 2)
plt.plot(z, k, color = 'green', label = 'regression line', linewidth = 3)
plt.ylim(-800000, 800000)
plt.xlabel('Predictions')
plt.ylabel(' Residuals')
plt.title('Residual Plot')
plt.legend()
```

```python
plt.figure(dpi = 200, figsize = (19, 8))
plt.hist(residual_table['Residuals'], color = 'red', bins = 200)
plt.xlabel('residuals')
plt.ylabel('frequency')
plt.title('Distribution of the error')
plt.show()
```

# Model coefficients

```python
coefficients_table = pd.DataFrame({'column' : x_train.columns, 'coefficients': lr.coef_})

coefficient_table = coefficients_table.sort_values(by = 'coefficients')
```

```python
plt.figure(dpi =100, figsize = (9, 8))
x = coefficient_table['column']
y = coefficient_table['coefficients']
plt.barh(x, y, color = 'green')
plt.xlabel('Coefficients')
plt.ylabel('Variable')
plt.title('Normalized coefficient Plot')
plt.show()
```
